Calcule el costo de envío de un paquete según su peso (en kg) y la zona de
destino. El usuario debe ingresar ambos valores.
Las zonas válidas son: local, regional, nacional.

Las zonas válidas son: local, regional, nacional.

Si la zona no es válida, debe indicar el error.

Ejemplo de salida esperada:

Ingrese la zona de destino (local/regional/nacional): regional
Costo de envío: $2500
Ingrese el peso del paquete (kg): 0.5
Ingrese la zona de destino (local/regional/nacional): internacional
Zona no válida. Las zonas disponibles son: local, regional, nacional.

```python
import sys
sys.path.append('../src')
from kg_zona import calcular_precio_kg

precios = {
    'hasta_1kg':    {'local': 500,  'regional': 1000, 'nacional': 2000},
    'entre_1_5kg':  {'local': 1000, 'regional': 2500, 'nacional': 4500},
    'mas_5kg':      {'local': 2000, 'regional': 5000, 'nacional': 8000},
}

peso_paquete = float(input('Ingrese el peso del paquete (kg): '))
zona_destino = input('Ingrese la zona de destino (local/regional/nacional): ')

precio_resultante = calcular_precio_kg(peso_paquete, precios, zona_destino)
if   precio_resultante == -1: 
    print('Zona no válida. Las zonas disponibles son: local, regional, nacional.')
else: 
    print(f'Costo de envío: ${precio_resultante} ' )

```
