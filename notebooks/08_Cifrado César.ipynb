Implemente el cifrado César, que consiste en desplazar cada letra del alfabeto
una cantidad fija de posiciones. El programa debe:
1. Solicitar un mensaje al usuario.
2. Solicitar un valor de desplazamiento (número entero).
3. Mostrar el mensaje cifrado.
4. Mostrar el mensaje descifrado a partir del cifrado (para verificar que
funciona).
Las letras deben rotar (después de la Z viene la A). Los caracteres que no sean
letras (espacios, signos de puntuación, números) se mantienen sin cambios. Debe
funcionar tanto con mayúsculas como con minúsculas, preservando el formato
original.

```python
import sys
sys.path.append('../src')
from cifrado_cesar import cesar 

mensaje = input('Ingrese un mensaje: ')
desplazamiento = int(input('Ingrese la cantidad del desplazamiento'))
copy = list(mensaje)

cifrado = cesar(copy, desplazamiento, cifrar= True)
descifrado = cesar(copy, desplazamiento, cifrar= False)

print (f'Mensaje cifrado: {cifrado} \nMensaje descifrado: {descifrado}')

```
