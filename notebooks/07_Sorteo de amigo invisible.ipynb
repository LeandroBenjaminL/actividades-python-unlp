Solicite al usuario una lista de nombres de participantes (separados por
coma). El programa debe asignar aleatoriamente a cada persona un amigo
invisible, cumpliendo que nadie se tenga a sí mismo.
Validaciones:
• Debe haber al menos 3 participantes.
• No debe haber nombres duplicados (comparar ignorando mayúsculas y
espacios extra).
Ejemplo de entrad

```python
import random 

ingresar_participantes = input('Ingrese los participantes separados por coma.  Ej: juan, cami, pedro')

participantes = ingresar_participantes.split(',')

duplicados = {p.strip().lower() for p in participantes}

if len(participantes) >= 3:
    if not (len(duplicados) < len(participantes)) :
        random.shuffle(participantes)
        for x in range(len(participantes)): 
            print(f'{participantes[x].strip().lower()} -> {participantes[(x+1) % len(participantes)].strip().lower()}')
    else:
        print('Hay nombres duplicados.')
else:
    print('Debia ingresar al menos 3 participantes')
```
